```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot(prediction: str, filename: str, title: str):
    df = pd.read_csv(prediction)
    
    plt.title(title)
    
    x = df['Actual']
    y = df['Predicted']
    
    sns.set_style('white-grid')
    sns.regplot(x=x, y=y)
    
    plt.savefig(f'./PICTURES/{filename}.png', bbox_inches='tight', dpi=100)
    
```

```python
# XGBOOST
plot('./PREDICTIONS/xg_boost_predictions.csv', 'xg_boost', 'XGBOOST')
```

```python
# RANDOM FOREST REGRESSOR
plot('./PREDICTIONS/random_forest_predictions.csv', 'rfr', 'RANDOM FOREST')
```

```python
# SUPPORT VECTOR 
plot('./PREDICTIONS/svm_predictions.csv', 'svm', 'SVM')
```

```python
# RANDOM FOREST
plot('./PREDICTIONS/rvm_predictions.csv', 'rvm', 'RVM')
```
